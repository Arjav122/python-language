Encapsulation means wrapping data (variables) and methods into a single unit (class) and controlling access to data.

👉 It is used for data protection and better code structure.

A public variable can be accessed from anywhere (inside or outside the class).

Syntax
self.variable

```python
class student:
    def __init__(self,name):
        self.name = name
```

```python
s = student("sudh")
```

```python
s.name
```

Protected Variable

A protected variable should be accessed only within the class and its child classes.
(In Python, this is a convention, not strict rule.)

Syntax
self._variable

```python
class student:
    def __init__(self,name):
        self._name = name
```

```python
s1 = student("sudh")
```

```python
s1._name
```

Private Variable

A private variable can be accessed only inside the same class.
Python uses name mangling to protect it.

Syntax
self.__variable

```python
class student:
    def __init__(self,name):
        self.__name = name
```

```python
s2 = student("sudh")
```

Name Mangling is a process where Python changes the name of a private variable internally to avoid accidental access or overriding.

It is applied to private variables (those starting with __ ). 

To access it _ClassName__var


```python
s2._student__name
```

```python
class student:
    def __init__(self,name):
        self.__name = name
        
    def get_name(self):
        return self.__name
```

```python
s = student("sudh")
```

```python
s.get_name()
```
