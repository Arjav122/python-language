Closure in Python

A closure is a function that remembers and uses variables from its outer function even after the outer function has finished execution.

```python
def outer():
    x = 10
    def inner():
        print(x)
    return inner
```

```python
closure = outer()
```

```python
def outer():
    x = 10
    def inner():
        print(x)
    def inner1():
        print(x)
    def inner2():
        print(x)
    return inner, inner1, inner2
```

```python
closure = outer()
```
