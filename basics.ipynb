Floating-Point Precision Issue in Python

A floating-point precision issue happens because decimal numbers cannot always be represented exactly in binary (base-2), which computers use.

Why does this happen?
Computers store numbers in binary
Some decimals like 0.1, 0.2 have infinite binary representations
So Python stores an approximate value
Just like 1/3 = 0.3333... in decimal.

```python
a = 0.3
b = 0.1 + 0.2

print(a == b)   # False
```

```python
print(0.1 + 0.2)
```

To solve this issue as well as other many functionalities we import math The math module provides mathematical functions and constants.

```python
import math
```

```python
print(math.sqrt(16))     
print(math.pow(2, 3))   
print(math.pi)  
print(math.e)   
print(math.ceil(4.2))    
print(math.floor(4.8))  
print(math.sin(90))  
print(math.cos(0))  
print(math.log(1))
```
