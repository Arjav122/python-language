enumerate() is used to loop over a sequence and get both index and value together.

enumerate(iterable, start=0)


```python
enumerate("euron")
```

```python
list(enumerate("euron"))
```

```python
fruits = ["apple", "banana", "cherry"]

for index, data in enumerate(fruits):
    print(index, data)
```

```python
fruits = ["apple", "banana", "cherry"]

for index, data in enumerate(fruits, start = 1):   # give start index by default 0
    print(index, data)
```

```python
s = "Python"
for i, ch in enumerate(s):
    print(i, ch)
```

zip() in Python

zip() is used to combine multiple iterables element-by-element.

zip(iterable1, iterable2, ...)


```python
names = ["A", "B", "C"]
marks = [90, 80, 70]

for n, m in zip(names, marks):
    print(n, m)
```

*zip() stops at the shortest iterable 

```python
a = [1, 2, 3, 4]
b = [10, 20]

list(zip(a, b))
```

More than two iterables

```python
names = ["A", "B"]
marks = [90, 80]
grades = ["A+", "B+"]

print(list(zip(names, marks, grades)))
```

Unzipping (reverse of zip)

```python
pairs = [(1, 'a'), (2, 'b')]

x, y = zip(*pairs)
print(x)
print(y)
```

reversed() is used to iterate over elements in reverse order.

reversed(iterable)


```python
nums = [1, 2, 3, 4]

for x in reversed(nums):
    print(x)
```

```python
reversed("mango")    # This is a reversed object, not yet converted or used.
```

```python
list(reversed("mango"))  # list(...) → converts that iterator object into a list
```

sorted() is used to return a new sorted list from any iterable.

sorted(iterable, key=None, reverse=False)

```python
nums = [3, 1, 4, 2]
sorted(nums)
```

```python
sorted(nums, reverse=True)
```
