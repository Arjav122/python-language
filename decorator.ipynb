A decorator is a function that adds extra behavior to another function without changing its code.

```python
def greet():
    print("Hello World")
```

```python
greet()
```

```python
def deco(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")
    return wrapper
```

```python
@deco
def greet():
    print("Hello World")
```

```python
greet
```

```python
greet()
```

```python
import time
def print_list(l):
    l1=[]
    start_time = time.time()
    for i in l:
        if i%2==0:
            l1.append(i)
        time.sleep(1)
    end_time = time.time()
    total_time = end_time - start_time
    return l1, total_time
```

```python
print_list([4, 5, 6, 6, 7, 8, 9, 67, 7, 89, 5, 41, 45, 7, 5, 46, 22, 5, 89])
```

```python
import time
def time_deco(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        print(start)
        output = func(*args, **kwargs)
        end = time.time()
        print(end)
        total_time = end-start
        return total_time, output
    return wrapper

```

```python
@time_deco
def print_list(l):
    l1 = []
    for i in l:
        if i%2==0:
            l1.append(i)
    return l1
```

```python
print_list([4, 5, 6, 6, 7, 8, 9, 67, 7, 89, 5, 41, 45, 7, 5, 46, 22, 5, 89])
```
