File Handling in Python (Read, Write, Append)
File handling is used to store data permanently in files.

Mode	Meaning

"r"	Read

"w"	Write (overwrite)

"a"	Append

"r+"	Read + Write

"w+"	Write + Read

"a+"	Append + Read

```python
open("test1.txt")
```

```python
f = open("test1.txt", 'w')
```

```python
f.write("Hello World")
```

```python
f.close()
```

```python
f = open("test1.txt")
```

```python
f.read()
```

```python
f.close()
```

```python
f = open("test1.txt", 'w')
```

```python
f.write("This world is best")
```

```python
f.close()
```

```python
f = open("test1.txt", 'a')
```

```python
f.write("\nThis data is appended")
```

```python
f.close()
```

```python
f = open("test1.txt", 'r')
```

```python
f.read()
```

```python
f = open("test1.txt", 'r')
for i in f:
    print(i.strip())
```

```python
f = open("test1.txt", 'r')
f.read(5)
```
