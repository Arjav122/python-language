seek() and tell()

These functions are used to control and check the file pointer position.

📍 File Pointer

When a file is opened, Python keeps a file pointer
→ it shows where reading or writing will happen next.

1️⃣ tell()

tell() returns the current position of the file pointer.

2️⃣ seek()

seek() is used to move the file pointer to a specific position.
