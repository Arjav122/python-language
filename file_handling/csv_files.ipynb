```python
import csv
```

```python
with open('sample.csv','r') as f :
    print(f)
    print(csv.reader(f))
    reader = csv.reader(f)
    for i in reader:
        print(i)
```

```python
with open('sample.csv','r') as f :
    reader = csv.reader(f)
    next(reader)
    for i in reader:
        print(i)
```

```python
with open('output.csv','w') as f :
    writer = csv.writer(f)
    writer.writerow(['sudhanshu','kumar'])
    writer.writerows([['data scince' , 'big data'],['tech','non tech'],['python','javascript']])
```
