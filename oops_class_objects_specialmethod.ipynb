What is __init__?

__init__ is a special method (constructor) that runs automatically when an object of a class is created.

```python
class Student:
    def __init__(self, name, age):    # self refers to the current object of the class
        self.name = name
        self.age = age

    def display(self):
        print(self.name, self.age)
```

Object - An object is an instance (real-world entity) created from a class.

```python
s1 = Student("Arjav", 20)
s2 = Student("Rahul", 21)
```

```python
s1.display()
```

__str__ is a special method that defines the human-readable string representation of an object.


(Without __str__)

```python
class Student:
    def __init__(self, name):
        self.name = name

s = Student("Arjav")
print(s)

```


(With __str__)

```python
class Student:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Student name is {self.name}"

s = Student("Arjav")
print(s)
```

__eq__ is a special (magic) method that defines how the == operator works for objects.

```python
class course :
    def __init__(self,title,mentor):
        self.title = title
        self.mentor= mentor
        
    def __str__(self):
        return f"my name is sudhanshu {self.title} , {self.mentor}"
    
    def __eq__(self, value):
        return self.title == value.title
```

```python
co1 = course("python","mentor")
co2 = course("pythonone","mentor")
```

```python
co1==co2
```

len() returns the number of elements in an object.

```python
class course :
    def __init__(self,title,mentor):
        self.title = title
        self.mentor= mentor
        
    def __str__(self):
        return f"my name is sudhanshu {self.title} , {self.mentor}"
    
    def __eq__(self, value):
        return self.title == value.title
    
    def __len__(self):
        return len(self.title) + len(self.mentor)
```

```python
a = course(["python","data","big data","data science"],"sudhanshu")
```

```python
print(len(a.title))
```

__call__ is a special method that allows an object to be called like a function.

```python
class course :
    def __init__(self,title,mentor):
        self.title = title
        self.mentor= mentor
        
    def __str__(self):
        return f"my name is sudhanshu {self.title} , {self.mentor}"
    
    def __eq__(self, value):
        return self.title == value.title
    
    def __len__(self):
        return len(self.title) + len(self.mentor)
    
    def __del__(self):
        print("i will delete the object")
        
    def __call__(self):
        return f"{self.title},{self.mentor}"
```

```python
a= course("python","sudh")
```

```python
a.mentor
```

```python
a.title
```

```python
a()
```
