✔ Keys are unique

✔ Values can be duplicated

✔ Unordered (conceptually)

✔ Mutable

❌ Keys must be immutable (no list/set)

as keys of dict are immutable so we cant use list or set or special case as a key but can use tuple string frozenset etc

```python
d1 = {123 : "abc"}   # int as a key allowed
```

```python
d1 = {"abc" : "abc"}  # string as a key allowed
```

```python
d1 = {(1, 2, 3) : "fsd6456"}  # tuple as a key allowed
```

```python
d1 = {[1, 2, 3] : "vgs"}  # list as a key not allowed same as set but frozenset is allowed
```

Looping dictionary

```python
for i in d.keys():
    print(i)
```

```python
for j in d.values():
    print(j)
```

```python
for i, j in d.items():
    print(i, j)
```
