Convert iterable → list (VERY IMPORTANT)

```python
print(list("mango"))             # ['m','a','n','g','o']
print(list(range(5)))            # [0,1,2,3,4]
print(list(reversed("mango")))   # ['o','g','n','a','m']
print(list(zip([1,2],[3,4])))    # [(1,3),(2,4)]
```

* Properly see difference between assignment vs shallow copy vs deep copy

b = s (Assignment)
Meaning

No copy is created.
Both variables point to the same object.

❌ No new list

Same memory location

Change in b affects s and vice-versa

s is b → True

```python
s = [1, 2, 3]
b = s
b.append(4)
print(s)

print(s is b)
```

Shallow Copy (copy())

Meaning
New outer list is created,
inner objects are shared.
✔ New outer list

❌ Inner lists are shared

Changes to nested objects affect both

Uses less memory

s is b → False



```python
s = [[1, 2], [3, 4]]
b = s.copy()
b[0].append(99)

print(s)

print(s is b)
print(s[0] is b[0])
```

Deep Copy (deepcopy())
Meaning

Completely independent copy.
Both outer and inner objects are copied.

✔ Everything is copied

✔ No shared objects

Changes do NOT affect original

Uses more memory

Safest for nested structures

```python
import copy

s = [[1, 2], [3, 4]]
b = copy.deepcopy(s)
b[0].append(99)
print(s)


print(s is b)
print(s[0] is b[0])
```
