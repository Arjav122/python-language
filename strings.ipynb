String Slicing means extracting a part of a string. 
string[start : end : step]
start → starting index (inclusive)
end → ending index (exclusive)
step → gap between characters

```python
print(s[1:4])   # From index to index
print(s[:4])    # From start
print(s[1:])    # Till end
print(s[-4:-2]) # Negative slicing
print(s[::2])   # Step slicing

print(s[0:100]) # if we give index out of range no problem i will give till last valid index

print(s[5:0])   # * here blank is answer as we are going last to first by default step size is +1 so conflict is there 

print(s[::1])
print(s[::-1])  # * Reverse string (very important)
```

String Interning in Python

String interning happens when Python automatically reuses a single memory object for identical, simple string literals to save memory.

Python thinks:
“Why create two separate "dog" strings?
One is enough.”
So Python creates only one "dog" in memory and both a and b point to it.
That sharing of the same string is called string interning.

String with space, Python does not always intern strings with spaces.


```python
a = "dog"
b = "dog"
```

```python
print(id(a))
print(id(b))
```

```python
a = "cat pig"
b = "cat pig"
```

```python
print(id(a))
print(id(b))
```

String Templating:

The string.Template class provides a way to create templates with placeholders.
Placeholders are identified by a dollar sign ($).
The substitute() method replaces placeholders with values.
Templating is useful for generating personalized emails or other documents.

```python
from string import Template
```

```python
t = Template("my name is $Name, my age is $Age")
```

```python
print(t.substitute(Name="Arjav", Age="22"))
```
