A while loop repeats a block of code as long as a condition is True.

while condition:
    # code to repeat

```python
i = 1
while i <= 5:
    print(i, end=" ")
    i += 1
```

break (stop loop)

```python
i = 1
while True:
    if i == 4:
        break
    print(i)
    i += 1
```

continue (skip current iteration)

```python
i = 0
while i < 5:
    i += 1
    if i == 3:
        continue
    print(i)
```

```python
s = "pineapple"
i=0
while i < len(s):
    print(s[i], end=" - ")
    i += 1
```
