A set is an unordered, mutable collection of unique elements.

❌ No duplicate values

❌ Unordered (no indexing)

✔ Mutable (can add/remove)

✔ Fast membership checking

❌ Cannot contain mutable items (list, set)

s1.difference(s2) - Returns a new set containing elements that are in s1 but NOT in s2.

```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5}

print(s1.difference(s2))
print(s2.difference(s1))
```

s1.intersection(s2) - 
Returns a new set containing elements that are common to both sets.

```python
print(s1.intersection(s2))
```

s1.union(s2)
Meaning
Returns a new set containing all unique elements from both sets.

```python
print(s1.union(s2))
```

s.pop()

s1.issubset(s2)
Meaning
Checks whether all elements of s1 are present in s2.

```python
s3 = {1, 2, 5}
s4 = {1, 2, 3, 5}

print(s3.issubset(s4))
print(s4.issubset(s3))
```
