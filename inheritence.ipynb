inheritance in Python

Inheritance means one class (child) uses the properties and methods of another class (parent).
It helps in code reuse and hierarchical design.

Single Inheritance - One parent → one child

```python
class Animal:
    def eat(self):
        print("Animal eats")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.eat()    # inherited method
d.bark()   # child method
```

Multilevel Inheritance - Grandparent → Parent → Child

```python
class A:
    def showA(self):
        print("A")

class B(A):
    def showB(self):
        print("B")

class C(B):
    def showC(self):
        print("C")

c = C()
c.showA()
```

Multiple Inheritance

One child → multiple parents

```python
class A:
    def showA(self):
        print("A")

class B:
    def showB(self):
        print("B")

class C(A, B):
    pass

c = C()
c.showA()
c.showB()
```

```python
class mentor:
    def __init__(self, name):
        self.name =  name
    def show(self):
        print(f"mentor {self.name}")
```

```python
class EuronMentor(mentor):
    def assign_course(self, course):
        print(f"{self.name}, {course}")
```

Method Overriding means
👉 Child class provides its own implementation of a method that already exists in the parent class.

The method name and parameters must be the same.

```python
class Parent:
    def show(self):
        print("This is Parent")

class Child(Parent):
    def show(self):
        print("This is Child")

obj = Child()
obj.show()
```

super() is used to call methods or constructor of the parent class from the child class.

```python
class Parent:
    def show(self):
        print("Parent method")

class Child(Parent):
    def show(self):
        super().show()
        print("Child method")

obj = Child()
obj.show()
```

MRO (Method Resolution Order) in Python – Notes

MRO defines the order in which Python searches for a method or attribute when it is called on an object.

👉 It is mainly important in multiple inheritance.

```python
class A:
    def show(self):
        print("A")

class B(A):
    pass

class C(B):
    pass

print(C.mro())
```

```python
class A:
    def show(self):
        print("A")

class B(A):
    def show(self):
        super().show()
        print("B")

class C(A):
    def show(self):
        super().show()
        print("C")

class D(B, C):
    def show(self):
        super().show()
        print("D")

d = D()
d.show()
```
