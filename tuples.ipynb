A tuple is an ordered and immutable collection of elements.

Tuples are used when data should not be changed after creation.

✔ Ordered (indexing allowed)

❌ Immutable (cannot change)

✔ Allows duplicate values

✔ Can store mixed data types

✔ Faster than list (slightly)

list are mutable and tuples are immutable major difference

```python
t = ()
type(t)
```

```python
t = (1, 2, 3, "apple", 5.561, False)
```

indexing and slicing same as list or strings

```python
print(t[0])
print(t[-2])
print(t[2:])
print(t[::-1])
```

in built functions

count() - Returns how many times a value appears in a sequence.

```python
a = (1, 2, 2, 3, 2)
print(a.count(2))
```

index() - Returns the index of the first occurrence of a value.

```python
z = (10, 20, 30, 20)
print(z.index(20))
```
