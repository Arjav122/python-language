for loop

A for loop is used to repeat a block of code for each item in a sequence.

```python
s = "watermelon"
for i in s:
    print(i.upper())
```

```python
list = [5, 6, 7, 8, 9]
for i in list:
    print(i)
```

The range() function is used to generate a sequence of numbers, commonly used with for loops.

range(start, stop, step)

start → starting number (default = 0)
stop → ending number (not included)
step → gap between numbers (default = 1)

```python
for i in range(5):
    print(i, end=" ")   # end="" → prevents line break
```

```python
for i in range(1, 10, 2):
    print(i)
```

```python
n = 5
for i in range(1, n+1):
    print("* " * i)
```

```python
n = 5
for i in range(1, n+1):
    for j in range(1, i+1):
        print(j, end=" ")
    print()
```
