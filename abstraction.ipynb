Abstraction means showing only essential details and hiding implementation details from the user.

How Abstraction is Achieved in Python

Abstract Base Classes (ABC)

@abstractmethod decorator

Module used:

from abc import ABC, abstractmethod

```python
from abc import ABC , abstractmethod
```

```python
class mentor(ABC):
    @abstractmethod
    def teach(self):
        pass
    
    @abstractmethod
    def mentor(self):
        pass
    
    @abstractmethod
    def course(self):
        pass
```

```python
class euronmentor(mentor):
    def teach(self):
        return "this will return teaching"
    
    def mentor(self):
        return "this will give a detaisl about mentor"
    
    def course(self):
        return "this will give a detaisl about cousre "
        
    
```

```python
eu = euronmentor()
eu.teach()
```

```python
m = mentor()
```

```python
class mentor(ABC):
    @abstractmethod
    def teach(self):
        pass
    
    @abstractmethod
    def mentor(self):
        pass
    
    @abstractmethod
    def course(self):
        pass
    
    def welcome(self):
        return "welcome to euron"
```

```python
m = mentor()
```
