List comprehension is a short and clean way to create a list in one line instead of using loops.

[expression for item in iterable if condition]

```python
[x*x for x in range(6)]
```


Conditional List Comprehension: Adding conditions to filter data during list creation

```python
[i for i in range(1, 21) if i%2==0]
```

```python
l = ["mango", "apple", "banana"]
[i.upper() for i in l]
```

Nested Loops in Comprehension:

```python
l = [[4, 8], [2, 7]]
[j for i in l for j in i]
```

Set Comprehension: Similar to list comprehension but creates sets, ensuring unique elements. It uses curly braces {}

```python
{x*x for x in range(1, 6)}
```

```python
{i for i in [1, 2, 2, 5, 4, 4, 8] if i%2 != 0}
```

```python
l = ["mango", "apple", "banana"]
{i[0] for i in l}
```

Dictionary comprehension is a short and clean way to create a dictionary using a single line of code.

{key_expression : value_expression for item in iterable if condition}


```python
{x: x**2 for x in range(1, 10)}
```

```python
d = {x: "even" if x % 2 == 0 else "odd" for x in range(1, 6)}
d
```
