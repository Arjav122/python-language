Conditionals in Python

Conditionals help a program decide what to do based on a condition.


if-else

If condition is True → if block runs
Else → else block runs

```python
age = 16
if age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible")
```

always remember this input() function take input as a string(str) that's why typecase to int

```python
num = int(input("enter number: "))

if num % 2 == 0:
    print("Even", num)
else:
    print("Odd", num)
```

if–elif–else is used when you have more than two conditions and only one block should run.

```python
marks = 85

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
```

A nested if–else means one if–else inside another if or else.

```python
username = "admin"
password = "1234"

if username == "admin":
    if password == "1234":
        print("Login successful")
    else:
        print("Wrong password")
else:
    print("Wrong username")
```

The ternary operator is a short way to write if–else in one line.

syntax - value_if_true if condition else value_if_false


```python
num = 5
"Even" if num % 2 == 0 else "Odd"

```

```python
a = 10
b = 20

max_val = a if a > b else b
print(max_val)
```

match–case is used to compare a value with different patterns, similar to switch-case in other languages.

```python
day = "wed"

match day:
    case "Mon":
        print("Monday")
    case "Tue":
        print("Tuesday")
    case _:
        print("its is just an another day")
```
