Multiple values returned from a function are stored as a tuple by default.

```python
def test2():
    return "sudh", 1, 5, 2.5, [1, 5, 6]
x = test2()
print(x)
```

Docstring in Python - A docstring (documentation string) is a string used to describe what a function, class, or module does.

Written using triple quotes """ """

Placed just after function / class / module definition

```python
def add(a, b):
    """
    This function takes two numbers
    and returns their sum.
    """
    return a + b
```

```python
add(5,3)
```

Accessing Docstring

```python
print(add.__doc__)
```

Positional / keyword / default arguments

Positional Arguments 📌 Order matters - Values are passed in the same order as parameters.


```python
def add(a, b):
    return a + b

add(2, 3)   # a=2, b=3
```

Keyword Arguments 📌 Order does NOT matter - Arguments are passed using parameter names.

✔ More readable
✔ Avoid confusion

```python
add(b=3, a=22)
```

Default Arguments - Some parameters have default values.

```python
def power(a, b=2):
    return a ** b

power(3)      # 3² = 9
power(3, 3)   # 3³ = 27
```

*args (Variable-length Positional Arguments)

*args allows a function to accept any number of positional arguments. 📌 Stored as a tuple

```python
def total(*args):
    k = 0
    for i in args:
        k = k+i
    return k

total(1, 2, 3, 4, 5)
```

*args stores extra arguments as a tuple, so number * args tries to multiply an integer with a tuple,tuple repeats

```python
def multiply(number, *args):
    return number*args

multiply(2, 5,6,7)
```

**kwargs (Variable-length Keyword Arguments)

**kwargs allows a function to accept any number of keyword arguments. 📌 Stored as a dictionary

```python
def details(**kwargs):
    for i in kwargs.items():
        print(i)
    return kwargs

details(name="Arjav", age=22, email="abc@gmail.com")
```

Using *args and **kwargs together

```python
def mix(*args, **kwargs):
    return args, kwargs

mix(2, 4, 6, name="Arjav", age=22)
```


"*" makes parameters after it keyword-only, meaning they must be passed using keywords.

```python
def func(a, b, *, c, d):
    print(a, b, c, d)

func(1, 2, c=3, d=4)   # ✅
func(1, 2, 3, 4)      # ❌ error
```

Lambda Function (Python) - A lambda function is a small anonymous function written in one line using the lambda keyword.

lambda arguments : expression

```python
add = lambda a,b : a+b

add(4,5)
```

```python
square = lambda x: x*x
square(5)
```

```python
even_odd = lambda x: "even" if x % 2 == 0 else "odd"
even_odd(4)
```

```python
max = lambda a,b : a if a>=b else b
max(5, 7)
```
